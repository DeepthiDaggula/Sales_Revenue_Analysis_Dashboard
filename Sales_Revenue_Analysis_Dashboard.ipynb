IMPORT LIBARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

CREATE SAMPLE SALES DATA

```python
sales = {
    "Product": [
        "Laptop",
        "Phone",
        "Tablet",
        "Headphones",
        "Monitor",
        "Keyboard",
        "Mouse"
    ],

    "Sales": [
        25,
        60,
        30,
        75,
        20,
        40,
        55
    ],

    "Revenue": [
        1500000,
        1800000,
        750000,
        375000,
        500000,
        200000,
        165000
    ]
}

df = pd.DataFrame(sales)

df
```

KPI's

```python
total_sales = df["Sales"].sum()
total_revenue = df["Revenue"].sum()

top_product = df.loc[df["Revenue"].idxmax()]

print("Total Sales :", total_sales)

print("Total Revenue : ₹", total_revenue)

print("\nTop Product")

print(top_product)
```

REVENUE BAR CHART

```python
plt.figure(figsize=(8,5))

plt.bar(df["Product"], df["Revenue"])

plt.title("Revenue by Product")

plt.xlabel("Products")

plt.ylabel("Revenue")

plt.xticks(rotation=30)

plt.show()
```

SALES PIE CHART

```python
plt.figure(figsize=(6,6))

plt.pie(
    df["Sales"],
    labels=df["Product"],
    autopct="%1.1f%%"
)

plt.title("Sales Distribution")

plt.show()
```

BUSINESS INSIGHTS

```python
print("Business Insights")

print("----------------------------")

print("Top Performing Product :", top_product["Product"])

print("Highest Revenue : ₹", top_product["Revenue"])

print("Overall Revenue : ₹", total_revenue)

print("Total Units Sold :", total_sales)
```
